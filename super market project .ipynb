```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("SuperMarket Analysis.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
if 'date' in df.columns:
    df['date']=df.pd_datetime(df['date'])
```

```python
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
```

```python
df.head()
```

```python
sales_by_product = df.groupby("Product line")["Sales"].sum().sort_values(ascending=False)
```

```python
print("\nSales by Product Line:\n", sales_by_product)
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x=sales_by_product.index, y=sales_by_product.values, palette="Blues_d")
plt.title("Total Sales by Product Line")
plt.xticks(rotation=90)
plt.ylabel("Sales")
plt.show()
```

Sales by Branch

```python
branch_sales = df.groupby("Branch")["Sales"].sum()
```

```python
print("\nSales by Branch:\n", branch_sales)
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x=branch_sales.index, y=branch_sales.values, palette="Oranges_d")
plt.show()
plt.title("Total Sales by Branch")
plt.ylabel("Sales")
```

```python
df.info()
```

 Customer Type Spending

```python
cust_sales = df.groupby("Customer type")["Sales"].sum()
```

```python
print("n\Customer Type Spending:\n", cust_sales)
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x=cust_sales.index, y=cust_sales.values, palette="viridis")
plt.show()
plt.title('Customer Type Spending')
plt.ylabel('sales')
```

Gender-wise Spending

```python
gender_wise=df.groupby('Gender')['Sales'].sum()
```

```python
print("\nGender-wise:\n", gender_wise)
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x=gender_wise.index, y=gender_wise.values, palette="coolwarm")
plt.title("Gender-wise Spending")
plt.ylabel("Total sales")
plt.show()
```

Monthly Sales Trend

```python
monthly_sales = df.groupby(df["Date"].dt.to_period("M"))["Sales"].sum()
```

```python
print("\nmonthly_sales:\n", monthly_sales)
```

```python
fig, ax1 = plt.subplots(figsize=(10,6))
ax1.bar(monthly_sales.index.astype(str), monthly_sales.values, color="skyblue", label="Monthly Sales")
ax1.set_xlabel("Month")
ax1.set_ylabel("Total Sales", color="blue")
ax1.plot(monthly_sales.index.astype(str), monthly_sales.values, color="red", marker="o", label="Trend Line")

plt.title("Monthly Sales (Bar + Line)", fontsize=12)
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

Payment Method Analysis

```python
payment_sales = df.groupby("Payment")["Sales"].sum()
```

```python
print("\nPayment Method Sales:\n", payment_sales)
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x=payment_sales.index, y=payment_sales.values, palette="Set2")
plt.title("Sales by Payment Method")
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Rating"], bins=15, kde=True, color="purple")
plt.title("Customer Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
payment_counts = df['Payment'].value_counts()
```
